```python
face_detector = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
plt.imshow(jin)
```

```python
face_zone = face_detector.detectMultiScale(jin)
# 前两个数据是人脸左上角的坐标
# 后两个数，宽度和高度
face_zone
```

```python
plt.imshow(dog)
```

```python
dog_head = dog[20:450,250:680]
plt.imshow(dog_head)
```

```python
dog_head2 = cv2.resize(dog_head,(73,73))
plt.imshow(dog_head2)
```

```python
plt.imshow(jin)
```

```python
for x,y,w,h in face_zone:
    jin[y:y+h,x:x+w] = dog_head2
    
plt.imshow(jin)
```

```python
import numpy as np
```

```python
jin = plt.imread('./jinzhengen.jpg')

jin = jin.copy()

for x,y,w,h in face_zone:
    
#     给定条件，圆的方程
#     半径，73/2
    r = int(w/2)
    cirle_point = np.array([w/2,h/2])
    for i in range(w):
        for j in range(h):
            
            if (i - cirle_point[0])**2 + (j - cirle_point[1])**2 < r**2:
                jin[y+i,x+j] = dog_head2[i,j]
            
plt.imshow(jin)
```

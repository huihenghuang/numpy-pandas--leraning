## 1. 使用pd.concat()级联

pandas使用pd.concat函数，与np.concatenate函数类似，只是多了一些参数：
```
pd.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
```

### 1)  简单级联

和np.concatenate一样，优先增加行数（默认axis=0）

````python
df1 = DataFrame(np.random.randint(0,150,size = (4,3)),columns=['Python','Math','En'],index=list('abcd'))
df2 = DataFrame(np.random.randint(0,150,size = (5,3)),columns=['Python','Math','En'],index=list('efghi'))
display(df1,df2)
````

````python
pd.concat([df1,df2])
````

````python
df3 = DataFrame(np.random.randint(0,150,size=(4,3)),columns=['Chinese','物理','化学'],index = list('abcd'))
df3
````

````python
pd.concat([df1,df3])
````

可以通过设置axis来改变级联方向

````python
pd.concat((df1,df3),axis = 1)
````

注意index在级联时可以重复

也可以选择忽略ignore_index，重新索引

````python
df4 = pd.concat([df1,df3],ignore_index=True)

# df4 重新设置索引
df4.index = list('abcdefgh')

df4
````

或者使用多层索引 keys  

concat([x,y],keys=['x','y'])

````python
# 期中
df1 = DataFrame(np.random.randint(0,150,size = (4,3)),columns=['Python','Math','En'],index = list('ABCD'))

# 期末
df2 = DataFrame(np.random.randint(0,150,size = (4,3)),columns=['Python','Math','En'],index = list('ABCD'))

display(df1,df2)
````

````python
df3 = pd.concat([df1,df2],keys = ['期中','期末'])
df3
````

````python
df3.unstack(level=0).stack()
````

### 2) 不匹配级联

````python
df1
````

````python
df4 = DataFrame(np.random.randint(0,150,size = (4,3)),columns=['Python','Math','Chinese'],index = list('BCDE'))
df4
````

不匹配指的是级联的维度的索引不一致。例如纵向级联时列索引不一致，横向级联时行索引不一致

有3种连接方式：

- 外连接：补NaN（默认模式）

````python
# outer 将所有的数据保存到新的DataFrame
pd.concat((df1,df4),join='outer')
````

- 内连接：只连接匹配的项

````python
# inner 将共同的列，进行合并
pd.concat((df1,df4),join = 'inner')
````

- 连接指定轴 join_axes

````python
df1.columns
````

````python
df4.columns
````

````python
# axis
pd.concat((df1,df4),join_axes=[df1.columns])
````

````python
# axis
pd.concat((df1,df4,df2),join_axes=[df4.columns])
````

````python
# axis
pd.concat((df1,df4),join_axes=df1.columns)
````

## 2. 使用pd.merge()合并

merge与concat的区别在于，merge需要依据某一共同的行或列来进行合并

使用pd.merge()合并时，会自动根据两者相同column名称的那一列，作为key来进行合并。

注意每一列元素的顺序不要求一致

````python
df_id = DataFrame({'id':[1,2,3,4]},index = list('ABCD'))
df_id
````

````python
df1
````

````python
df1 = pd.concat([df1,df_id],axis = 1)
df1
````

````python
df2 = DataFrame(np.random.randint(0,150,size=(4,3) ),index=list('ABCD'),columns=['语文','政治','历史'])
df2 = pd.concat([df2,df_id],axis = 1)
df2
````

###  1) 一对一合并

````python
# MergeError: No common columns to perform merge on. Merge options: left_on=None, right_on=None, left_index=False, right_index=False
df1.merge(df2)
````

````python
# merge 默认情况下，列增加了
df1.merge(df2)
````

````python
df2['id']['A'] = 10
df2['id']['D'] = 20
df2
````

````python
# merge进行融合的时候，需要有共同的一列
# 根据这一列中，共同的值，进行合并
df1.merge(df2)
````

### 2) 多对一合并

````python
df2['id']['A'] = 2
df2
````

### 4) key的规范化

- 使用on=显式指定哪一列为key,当有多个key相同时使用

````python
df2
````

````python
df3 = DataFrame(np.random.randint(0,150,size = (4,4)),columns=['语文','Python','Math','id'],index=list('ABCD'))
df3['id']['A'] = 110
df3
````

````python
# 指明on根据那一列进行融合，不指名，根据根据所有的共同的列名进行融合
# 不指名的时候，语文和id都是共同列
df2.merge(df3,on = 'id',suffixes=('_期中','_期末') )
````

````python
df4 = DataFrame(np.random.randint(0,150,size = (4,4)),columns=['En','Python','Math','Id'],index=list('ABCD'))
df4['Id']['A'] = 110
df4
````

````python
df2
````

````python
df2.merge(df4)
````

````python
# df2 左边
# df4 右边
df2.merge(df4,left_on='id',right_on='Id')
````

- 使用left_on和right_on指定左右两边的列作为key，当左右两边的key都不想等时使用

````python
df5 = DataFrame(np.random.randint(0,150,size = (4,3)),columns=['En','Python','Math'],index=[2,3,20,110])
df5
````

````python
df2
````

````python
df2.merge(df5,left_on='id',right_index=True)
````

## 3. 案例分析：美国各州人口数据分析

首先导入文件，并查看数据样本

````python
areas = pd.read_csv('./state-areas.csv')
# 类型是：DataFrame
areas
````

合并pop与abbrevs两个DataFrame，分别依据state/region列和abbreviation列来合并。

为了保留所有信息，使用外合并。

````python
pop2 = pop.merge(abbr,left_on='state/region',right_on='abbreviation',how = 'outer')
pop2.shape
````

去除含有缺失数据的行

查看数据是否缺失

找出2010年的全民人口数据,df.query(查询语句)

要点总结：

## 回顾：Series/DataFrame运算与ndarray运算的区别

- Series与DataFrame没有广播，如果对应index没有值，则记为NaN；或者使用add的fill_value来补缺失值
- ndarray有广播，通过重复已有值来计算
